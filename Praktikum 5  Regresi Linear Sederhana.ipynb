```python
x = X[['sqft_living']]
y = Y
```

```python
xg = x.values.reshape(-1,1)
```

```python
yg = y.values.reshape(-1,1)
```

```python
xg = np.concatenate((np.ones(len(x)).reshape(-1, 1), x), axis=1)
```

```python
def computeCost(x, y, theta):
  m = len(y)

  h_x = x.dot(theta)
  j = np.sum(np.square(h_x - y)) * (1 / (2 * m))

  return j
```

```python
def gradientDescent(x, y, theta, alpha, iteration):
  print("Running Gradient Descent...")

  j_hist = []
  m = len(y)

  for i in range(iteration):
    h_x = x.dot(theta)
    theta = theta - (alpha / m) * np.dot(x.T, (h_x - y))
    j_hist.append(computeCost(x, y, theta))

  return theta, j_hist
```

```python
def regresi_linier(x, y, theta, alpha, iteration):
  print("Running Gradient Descent...")

  j_hist = []
  m = len(y)

  for i in range(iteration):
    h_x = x.dot(theta)
    gradient = (1 / m) * x.T.dot(h_x - y)
    theta = theta - alpha * gradient
    j_hist.append(computeCost(x, y, theta))

  return theta, j_hist

```

```python
# Assuming xg and yg are your features and target variable respectively
theta = np.zeros((xg.shape[1], 1))  # Initializing theta with zeros
iterations = 2000
alpha = 0.001

def gradientDescent(x, y, theta, alpha, iterations):
    m = len(y)
    cost_history = []

    for _ in range(iterations):
        hypothesis = np.dot(x, theta)
        loss = hypothesis - y
        gradient = np.dot(x.T, loss) / m
        theta -= alpha * gradient
        cost = np.sum(loss ** 2) / (2 * m)
        cost_history.append(cost)

    return theta, cost_history

theta, cost = gradientDescent(xg, yg, theta, alpha, iterations)
print('Theta found by Gradient Descent: slope = {} and intercept = {}'.format(theta[1], theta[0]))
```
